Gera animação da simulação

```python
#Gera animação
fps = 4 
n = 150
passos = 60

m = np.zeros((n,n))
m[n//2+1,n//2] = 1
m[n//2-1,n//2] = 1
m[n//2,n//2+1] = 1
m[n//2,n//2-1] = 1

a = s.cos2(n)
cdif = 0.3
cmax = 6*a
cinit = cmax - 0.1*a

sim = s.Sim(m, cdif = cdif, cmax = cmax, cinit = cinit)

eixo = 'log'
pop = [sim.pop]
x = [0]
(fig, ax, a1, a2, a3) = s.grafico(pop, sim, eixo = eixo, passos = passos)
def init():
     pass
def animate(i):
    if i % 10 == 0:
        print(passos - i, end=' ')
    pop.append(sim.pop)    
    a1.set_data(sim.m)
    ax.set_title('Pop = ' + str(sim.pop) + ', t = ' + str(i))    
    a2.set_data(sim.c)
    if eixo == 'log':
        a3.set_data(x, np.log(pop))
    else:
        a3.set_data(x, pop)
    
    sim.r = 0.0 + (1-0.0)/(1+np.exp(1*(10-i)))
    #print(sim.r, sim.pop, end=' ')
    sim.att()
plt.close('all')
# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=passos, interval=1000/fps)
HTML(anim.to_html5_video())
```

Média de 'rs' simulações

Roda 'rs' simulações de n passos e calcula a média da população por passo

```python
ns = 5
rs = 3

n = 150
passos = 35

m = np.zeros((n,n))
m[n//2-1,n//2] = 1        
m[n//2+1,n//2] = 1 
m[n//2,n//2-1] = 1 
m[n//2,n//2+1] = 1

dados = np.zeros((ns, rs, passos))
dados_cmax = []
for k in range(ns):
    
    a = s.cos2(n)
    cdif = 0.2
    cmax = (4 + k)*a
    cinit = cmax - 0.1*a
    r_init = 0
    r_meio = 5
    r_a = 1
    
    dados_cmax.append(np.max(cmax))
    print('cmax = {:.2f}'.format(dados_cmax[-1]), end = '\t')
    for j in range(rs):  # Roda as 'rs' simulações
        print('Simulação:', j)

        sim = s.Sim(m, cdif = cdif, cmax = cmax, cinit = cinit)

        for i in range(passos): # Roda os n passos de cada simulação
            if i % 10 == 0:
                print(passos - i, end= ' ')
            dados[k, j, i] = sim.pop
            sim.r = sim.r = r_init + (1-r_init)/(1+np.exp(r_a*(r_meio-i)))
            sim.att()        
        print('pop =', sim.pop)
    
media = np.mean(dados, axis = 1)
if rs > 1:
    sigma = np.std(dados, axis = 1, ddof = 1)/np.sqrt(rs)

plt.close('all')
x = np.arange(passos)
plots = []
for i in range(ns):
    linha, = plt.plot(x, media[i], label = '{:.2f}'.format(dados_cmax[i]))
    plots.append(linha)
plt.legend(handles = plots)
plt.yscale('log')
plt.xlabel('t')
plt.ylabel('N')
plt.show()
```

```python
plt.close('all')
plots = []
for i in range(ns):
    linha = plt.errorbar(x, media[i], yerr = sigma[i], fmt = '.', label = '{:.2f}'.format(dados_cmax[i]))
    plots.append(linha)
    
    np.savetxt('k{:.0f}.txt'.format(dados_cmax[i]*100), np.array([x, np.log(media[i]), sigma[i]/media[i]]).T, delimiter = '\t')
plt.legend(handles = plots)
plt.yscale('log')
plt.xlabel('t')
plt.ylabel('N')
plt.show()
```
